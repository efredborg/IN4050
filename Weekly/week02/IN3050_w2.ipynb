### 1. Given the function $ f(x) = -x^{4} + 2x^{3} + 2x^{2} - x $ and it's derivative $f'(x) = {- 4 x^3 + 6 x^2 + 4 x - 1}$

```python
# Standard import and functions
# Run this cell first
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return -np.power(x, 4) + 2 * np.power(x, 3) + 2 * np.power(x, 2) - x

def df(x):
    return -4 * np.power(x, 3) + 6 * np.power(x, 2) + 4 * x - 1
```

#### a. Plot the function, and its gradient (derivative) from $x = -2$ to $x = 3$

```python
# Add your solution here
x = np.linspace(-2,3,100)
plt.figure(figsize=(8,8))
plt.plot(x, f(x), c='k', label=r'$f(x)$', ls='solid')
plt.plot(x, df(x), c='k',label=r'$\frac{df(x)}{dx}$', ls='dotted')
plt.legend()
plt.show()
```

#### b. Maximize using gradient ascent. You can try step size 0.1 and start somewhere in the range $[-2, 3]$. 

Gradient ascent: $x^{(k+1)} = x^{(k)} + \gamma \Delta f(x^{(k)})$

```python
# Add your solution here

tol = 1e-3
x0 = np.random.choice(x)
print(x0)
gamma = 0.1
optimum = x0
gradient=df(optimum)
n = 0
while gradient>tol:
    optimum += gamma*gradient
    gradient = df(optimum)
    n += 1
    if n>1e3:
        break
        
plt.figure(figsize=(8,8))
plt.plot(x, f(x), c='k', label=r'$f(x)$', ls='solid')
plt.plot(optimum, f(optimum), 'ro', label='Optimum')
plt.legend()
plt.show()
print(optimum)
    
    
    
```

The algorithm would not be able to find a any maximum if the starting point is at the local/global minumum. At that point the gradient would be close to zero and the algorithm would break from the progression loop. 


#### c. Assume that we are only interested in maxima of $f(x)$, where $-2 \leq x \leq 3$, and x increases in step length of 0.5 ($\Delta x = 0.5$). Perform an exhaustive search to maximize $f(x)$ and plot the result

```python
# Add your solution here
x0=-2
gamma = 0.5
optimum = x0
gradient = df(optimum)
N = int((3--2)/gamma+1)

for i in range(N):
    if f(x0+i*gamma)>f(optimum):
        optimum=x0+i*gamma
print(optimum)
        
```
